```python
import pandas as pd
import numpy as np

trdt1=pd.read_csv('/home/aaftab/Github/Computational-Systems-Biology-Project/DREAM5 SYSTEMS GENETICS CHALLENGE/PART B/TrainingGenotype.csv')
trdt=pd.read_csv('/home/aaftab/Github/Computational-Systems-Biology-Project/DREAM5 SYSTEMS GENETICS CHALLENGE/PART B/DREAM5_SysGenB_TrainingData/DREAM5_SysGenB_TrainingExpressionData.csv')
tstdt=pd.read_csv('/home/aaftab/Github/Computational-Systems-Biology-Project/DREAM5 SYSTEMS GENETICS CHALLENGE/PART B/DREAM5_SysGenB_TestData/DREAM5_SysGenB2_TestExpressionData.csv')

trdt.drop('A', axis=1, inplace=True)
trdt=np.array(trdt)
trdt=np.transpose(trdt)
trdt=pd.DataFrame(trdt)

tstdt.drop('A', axis=1, inplace=True)
tstdt=np.array(tstdt)
tstdt=np.transpose(tstdt)
tstdt=pd.DataFrame(tstdt)

features=trdt.copy()

testfeatures=tstdt.copy()

results1=trdt1['Phenotype1']
results2=trdt1['Phenotype2']

from sklearn import preprocessing
standard= preprocessing.StandardScaler()

features = standard.fit_transform(features)
testfeatures=standard.fit_transform(testfeatures)

features=pd.DataFrame(features)
testfeatures=pd.DataFrame(testfeatures)
```

```python
from sklearn import linear_model

lmr = linear_model.LinearRegression()
lgr2= linear_model.LinearRegression()

modellr = lmr.fit(features,results1)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, results1, test_size=0.3, random_state=0)
modelval=lgr2.fit(X_train,y_train)
y_pred = modelval.predict(X_test)
y_pred2=modelval.predict(X_train)

import math
from math import sqrt
from sklearn.metrics import mean_squared_error

mse1=mean_squared_error(y_train, y_pred2)
rmse1=sqrt(mse1)
print(str(rmse1)+'Training error')
mse2=mean_squared_error(y_test, y_pred)
rmse2=sqrt(mse1)
print(str(rmse2)+'Testing Error')

print ('score ' + str(modelval.score(X_train, y_train)))
print ('score '+str(modelval.score(X_test, y_test)))
        
predictions = modellr.predict(testfeatures)
pred=pd.DataFrame(predictions)
        
pred.to_csv('LinearRegressionPredictions1.csv')

from sklearn import linear_model

lmr = linear_model.LinearRegression()
lgr2= linear_model.LinearRegression()

modellr = lmr.fit(features,results2)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, results2, test_size=0.3, random_state=0)
modelval=lgr2.fit(X_train,y_train)
y_pred = modelval.predict(X_test)
y_pred2=modelval.predict(X_train)

import math
from math import sqrt
from sklearn.metrics import mean_squared_error

mse1=mean_squared_error(y_train, y_pred2)
rmse1=sqrt(mse1)
print(str(rmse1)+'Training error')
mse2=mean_squared_error(y_test, y_pred)
rmse2=sqrt(mse1)
print(str(rmse2)+'Testing Error')

print ('score ' + str(modelval.score(X_train, y_train)))
print ('score '+str(modelval.score(X_test, y_test)))
        
predictions = modellr.predict(testfeatures)
pred=pd.DataFrame(predictions)
        
pred.to_csv('LinearRegressionPredictions2.csv')
```
